```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('./datasets/Scotia_Business_Inc._Export_Development_and_Investment_Attraction_Activity.csv')

# Get the top 10 sectors based on total financial contribution
top_10_sectors = df.groupby('nsbi_sector')['NSBI Financial Contribution (Commitment/Announced)'].sum().nlargest(10)

# Plot the distribution of financial contribution by sector
top_10_sectors.plot(kind='bar', rot=45)
plt.xlabel('Sector')
plt.ylabel('Total Financial Contribution')
plt.title('Distribution of Financial Contribution by Top 10 Sectors')
plt.show()
```
