```python
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt


# load your data into a Pandas DataFrame
df = pd.read_csv('./datasets/Municipal_Fiscal_Statistics_-_Financial_Condition_Indicators.csv')

# select only the non-numeric columns from the data
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_data = df[categorical_cols]
print(categorical_data)
# create a LabelEncoder object to convert categorical data to numeric
le = LabelEncoder()

# apply the LabelEncoder to each categorical column
for col in categorical_cols:
    categorical_data[col] = le.fit_transform(categorical_data[col])

# replace the original categorical columns with the encoded data
df[categorical_cols] = categorical_data



# create a SimpleImputer object to replace missing values with the mean
imputer = SimpleImputer(strategy='most_frequent')


# apply the imputer to fill in the missing values
df_imputed = imputer.fit_transform(df)

# instantiate a PCA object with the number of components you want to keep
pca = PCA(n_components=2)

# fit the PCA model to the imputed data
pca.fit(df_imputed)

# transform the data using the PCA model
transformed = pca.transform(df_imputed)

# print(transformed)
```

```python
# create a scatter plot of the transformed data
plt.scatter(transformed[:, 0], transformed[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Scatter Plot')
plt.show()
```
