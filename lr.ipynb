```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Create a DataFrame with hardcoded values
data = pd.DataFrame({
    "program_funding": [100, 70, 120, 155, 135],
    "investment_attraction_activity": [25, 60, 30, 11, 22],
    "payroll_growth": [10, 12, 15, 14, 18]
})

# Step 2: Data preprocessing
# You may need to remove any missing or irrelevant data here

# Step 3: Perform linear regression
X = data[["program_funding", "investment_attraction_activity"]]
y = data["payroll_growth"]
reg = LinearRegression().fit(X, y)
print(reg.coef_)  # Print the coefficients of the linear regression

# Step 4: Visualize the data
plt.scatter(data["program_funding"], data["payroll_growth"])
plt.xlabel("Program Funding")
plt.ylabel("Payroll Growth")
plt.show()
```
