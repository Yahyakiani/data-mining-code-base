```python
import pandas as pd
from scipy.stats import uniform, beta
import numpy as np
import matplotlib.pyplot as plt

# Read in the CSV data
df = pd.read_csv('./datasets/Scotia_Business_Inc._Export_Development_and_Investment_Attraction_Activity.csv')

# Clean the data as necessary

# Modelling the likelihood of a company receiving a grant based on its sector using a beta distribution
# You can also use other probability distributions such as uniform, normal, etc. depending on your model
sector_probs = beta(a=2, b=5) # Example beta distribution with shape parameters a=2, b=5

# Generate a large number of samples from the model
n_samples = 2542
sector_samples = sector_probs.rvs(size=n_samples)
df_resampled = df.sample(n=len(df), replace=True)
# Use the generated samples to estimate the probability of a company receiving a grant based on its sector
criteria = df['nsbi_sector'] == 'ICT (includes Digital Media)'
prob_ict = np.mean(sector_samples[criteria])

# Visualize the results
plt.hist(sector_samples, bins=50, alpha=0.5)
plt.axvline(x=prob_ict, color='red', linestyle='--')
plt.xlabel('Probability of receiving a grant')
plt.ylabel('Number of Companies')
print(prob_ict,'ICT (includes Digital Media)')
plt.show()
```
